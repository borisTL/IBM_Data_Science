# Data Science Overview

## Data Science Task Categories

### 1. Data Management
- Storage, management, and retrieval of data.

### 2. Data Integration and Transformation
- Streamline data pipelines.
- Automate data processing tasks.

### 3. Data Visualization
- Provide graphical representation of data.
- Assist with communicating insights.

### 4. Modelling
- Building, Deployment, Monitoring, and Assessment of Data and Machine Learning models.

## Support for Data Science Tasks

### 1. Code Asset Management
- Store & manage code.
- Track changes.
- Allow collaborative development.

### 2. Data Asset Management
- Organize and manage data.
- Provide access control.
- Backup assets.

### 3. Development Environments
- Develop, test, and deploy code.

### 4. Execution Environments
- Provide computational resources.
- Run the code.

## Data Science Ecosystem

The ecosystem consists of many open source and commercial options. This includes both traditional desktop applications and server-based tools. There are also cloud-based services accessible via web-browsers and mobile interfaces.

### 1. Data Management Tools
- **Relational Databases**: MySQL, PostgreSQL (Open Source), IBM Db2, SQL Server (Commercial & Cloud services).
- **NoSQL Databases**: MongoDB, Apache Cassandra.
- **Big Data Platforms**: Apache Hadoop, Apache Spark.

### 2. Data Integration and Transformation Tools
- Apache Airflow, Apache Kafka.

### 3. Data Visualization Tools
- **Commercial Offerings**: Cognos Analytics, Tableau, PowerBI.
- Used for building dynamic and interactive dashboards.

### 4. Code Asset Management Tools
- **Git**: Essential for code asset management.
- **GitHub**: Web-based platform for storing and managing source code. Ideal for collaborative software development with features such as version control, issue tracking, and project management.

### 5. Development Environments
- **Jupyter Notebooks**: Provides an interactive environment for creating and sharing code, descriptive text, data visualizations in a web-browser interface.
- **RStudio**: IDE designed for R programming language, popular for statistical computing and data analysis.
